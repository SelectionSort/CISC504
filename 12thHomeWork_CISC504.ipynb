```python
# 3- Omit all rows/observations with missing 
# 'Total Project Cost' information.
dataset1.isnull().sum()
```

```python
dataset2 = dataset1[np.isfinite(dataset1['Total Project Cost'])]
```

```python
dataset2
```

```python
# No more NaN value for 'Total Project Cost'
dataset2.isnull().sum()
```

```python
# 4- Filter out the observations for the range of years 2014-2019
filter1 = (dataset2["Application Received Year"] >= 2014) & (dataset2["Application Received Year"] <= 2019)
dataset3 = dataset2[filter1]
dataset3
```

```python
# 5- Develop an informative plot of total project cost by facility type
g = sns.FacetGrid(dataset3, margin_titles=True)
g.map(plt.scatter, "Facility Type", "Total Project Cost", alpha=.7)
g.add_legend();
```

```python
# 6- Stratify the plot you developed by year
with sns.axes_style(style='ticks'):
    g = sns.catplot("Facility Type", "Total Project Cost", "Application Received Year", data=dataset3, kind="box")
```

```python
# 7- Develop an informative visual plot of your choice
g = sns.FacetGrid(dataset3, margin_titles=True)
g.map(plt.scatter, "Application Received Year", "Total Project Cost", alpha=.7)
g.add_legend();
g.set(xticks=[2014, 2015, 2016, 2017, 2018, 2019]);
```
